```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
#functional plot
plt.plot(x,y)
plt.xlabel('Xlabel')
plt.ylabel('Ylabel')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

       # Object oriented method

```python
fig=plt.figure()
axes=fig.add_axes([0.2,0.4,0.8,0.8])
axes.set_xlabel('Xlabel')
axes.set_ylabel('Ylabel')
axes.plot(x,y)
```

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes

fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].set_title('First plot')
axes[1].set_title('Second plot')
axes[0].set_xlabel('xlabel')
axes[1].set_ylabel('ylabel')
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#We can iterate through this array axes
for current_ax in axes:
    current_ax.plot(x,y)
    
```

```python
#Specifying figsize and DPI
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(12,3),dpi=100)
```

```python
#Legends, labels and titles

fig=plt.figure()
axes=fig.add_axes([0.2,0.2,0.8,0.8])
axes.plot(x,x**2,'r',label='square')
axes.plot(x,x**3,'b',label='cube')
axes.legend()
axes.set_title('TITLE')
axes.set_xlabel('X values')
axes.set_ylabel('Y values')
```

```python
#Setting colors, linewidths, linetypes
```

```python
fig,axes=plt.subplots()
axes.plot(x,x**2,color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
axes.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.plot(x,x**3)
```

```python
#Setting Xlimits and Y limit to control the graph
```

```python
fig,axes=plt.subplots()
axes.set_xlim([0,1])
axes.set_ylim([0,2])
axes.plot(x,x**3)
#make sure to pass a list
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed',linewidth=0.5)
```

```python
#twin axis
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
n = np.array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```
